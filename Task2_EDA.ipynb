```python
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.mean(numeric_only=True)
df.median(numeric_only=True)
df.std(numeric_only=True)
```

```python
df.hist(figsize=(12,10), bins=20)
plt.show()
```

```python
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Fare'])
plt.title("Fare Outliers")
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[['Survived','Age','Fare','Pclass']].dropna(), hue='Survived')
plt.show()
```

```python
fig = px.scatter(df, x="Age", y="Fare", color="Survived",
                 title="Age vs Fare Survival")
fig.show()
```

CONCLUSION
1. Passengers in higher class survived more.
2. Females survived more than males.
3. Higher fare passengers had higher survival rate.
4. Children had better survival probability.
5. Outliers exist in Fare column.
